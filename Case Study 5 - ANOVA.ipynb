```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
#Read the following dataset
df = pd.read_csv(r'Downloads\Sales_data_zone_wise.csv')
df.head()
```

```python
#Null value checking
df.isnull().sum()
```

```python
#There is no null value
```

```python
data = df.drop(["Month"], axis =1)
data.head()
```

```python
#Total sales generated by all zones in each month  -  #Row wise addition
sum_row=data.sum(axis=1)
sum_row
```

```python
#Frequency graph of the given data
freqgraph = data.select_dtypes(include=['int32','int64'])
freqgraph.hist(figsize=(15,15))
plt.show()
```

```python
#H0 = "All Zone Generate same amount of sales"
#Ha = "All Zone do not Generate same amount of sales"

#Significance level = 0.05
```

```python
#Anova F and P value using F_one way
fvalue,pvalue=stats.f_oneway(data["Zone - A"],data["Zone - B"],data["Zone - C"],data["Zone - D"])
print("F value is:",fvalue)
print("P value is:",pvalue)
if pvalue<0.05:
    print("Reject null hypothesis")
else:
    print("Accepts null hhypothesis")
    
```

```python
#From the given data it rejects null hypothesis and which means All Zones do not generate same amount of sales they all are different
```
